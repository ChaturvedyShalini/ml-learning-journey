LOGISTIC REGRESSION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("logistic_data.csv")
dataset.drop(columns=["EstimatedSalary"], inplace = True)
dataset.head()
```

```python
sns.scatterplot(x="Age", y = "Purchased", data = dataset)
plt.show()
```

The Age vs Purchased plot shows a clear binary separation. 
Younger users mostly belong to class 0, while older users belong to class 1.
This step-like transition aligns with the sigmoid behavior of Logistic Regression,
making it suitable for binary classification.


```python
x = dataset[['Age']]
y = dataset['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

`lr` is an object of the LogisticRegression model.
It is used to train the model on data and later make binary predictions (0 or 1).


```python
lr.score(x_test, y_test)*100
```

The Logistic Regression model achieved an accuracy of **93.75%** on the test dataset,
meaning it correctly predicted 93.75% of the purchase outcomes.


```python
lr.predict([[40]])
```

```python
sns.scatterplot(x="Age", y = "Purchased", data = dataset)
sns.lineplot(x = "Age", y = lr.predict(x), data= dataset, color = 'red')
plt.show()
```

The curve looks Z-shaped because only class predictions (0 and 1) are plotted.
Logistic Regression internally follows an S-shaped probability curve, but converting
probabilities into class labels creates a sharp transition in the graph.
