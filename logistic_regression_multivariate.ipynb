```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("logistic_data.csv")
dataset.head()
```

```python
sns.scatterplot(x="Age", y="EstimatedSalary", data=dataset)
plt.show()
```

Here, the `hue='Purchased'` parameter is used to color the data points based on purchase status.
This helps us visually distinguish between customers who purchased the product (1) and those who did not (0).


```python
sns.scatterplot(x="Age", y="EstimatedSalary", data=dataset, hue="Purchased")
plt.show()
```

From the scatter plot, we can see that the two classes (Purchased = 0 and Purchased = 1)
are mostly separated. This indicates that the data is suitable for binary classification
using Logistic Regression.


The legend is used to explain the meaning of different colors in the plot.
It helps us easily identify which points represent Purchased = 0 and Purchased = 1.


```python
sns.scatterplot(x="Age", y="EstimatedSalary", data=dataset, hue="Purchased")
plt.legend(loc=1)
plt.show()
```

The data is reasonably linearly separable, making it suitable for Logistic Regression.


```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test, y_test)*100   # 93.75% accuracy
```

```python
lr.predict([[22,19000]])  # means it is not purchased the result shows array([0])
```

Making decission boundary through which we can see that how it make separation between the classes.

To plot the decision boundary, we use the `mlxtend` library as it provides additional
functions that are not available in basic plotting libraries.


```python
lr.coef_
```

```python
lr.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr)
plt.show()
```

The decision boundary is based on the sigmoid function  
y = 1 / (1 + e^(-x)), where  
x = m1x1 + m2x2 + bias.

Here, m1 and m2 are the coefficients of the input features (x1 and x2),
and the bias represents the intercept term.


**Linear equation (z):**

z = 0.91969081x1 + 0.00145296x2  103.92949986

**Sigmoid (Logistic Regression):**

y = 1 / (1 + e^(z))

**Substitute z:**

y = 1 / (1 + e^((0.91969081x1 + 0.00145296x2  103.92949986)))

**Decision Boundary (classification line)**

Decision boundary tab hoti hai jab y = 0.5, yaani z = 0:

0.91969081x1 + 0.00145296x2  103.92949986 = 0

Ye line hi classes ko separate karti hai

**Explanation:**

Here, the coefficients represent the importance of each feature, and the intercept acts as the bias term that shifts the decision boundary.
